### Parse the Heroes' Attribute from the website (http://dota2.gamepedia.com/Table_of_hero_attributes)

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

### Get Heroes Features table

```python
url = 'http://dota2.gamepedia.com/Table_of_hero_attributes'
d2 = requests.get(url)
```

```python
soup = BeautifulSoup(d2.content)
```

```python
table = soup.find('table', attrs={'class':'wikitable'})
```

```python
headings = [th.get_text().rstrip() for th in table.find("tr").find_all("th")]
```

```python
datasets=[]
for row in table.find_all("tr")[1:]:
    datasets.append([td.get_text().rstrip() for td in row.find_all("td")])
```

```python
heroes_features = pd.DataFrame(datasets, columns=headings)
```
